```python
import numpy as np
import random
from sklearn.datasets.samples_generator import make_regression 
import pylab
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Getting Linear data from sklearn dataset with 1 independent feature and 1 Dependent feature
x, y = make_regression(n_samples=100, n_features=1, n_informative=1, random_state=0, noise=20) 
```

```python
# Finding Optimal weights that helps the Regression Line to fit data
def weight_update(x,y,lr=0.01,epochs=1000):
    m = 10
    c = 5
    n = len(x)
    ep = False
    mse = False
    converged = False
    epoch = 0
    while not converged:
        y_pred = m*x + c
        temp = y - y_pred
        m_g = (-2/n) * sum([x[i]*(y[i] - y_pred[i]) for i in range(len(x))])
        c_g = (-2/n) * sum([y[i] - y_pred[i] for i in range(len(x))])
        m = m - (lr*m_g)
        c = c - (lr*c_g)
        error = sum([(y[i]-(m*x[i] + c))**2 for i in range(len(x))])        
        if epoch==1000:
            ep = True
        if error <= 0.01:
            mse = True
        converged = ep or mse
        epoch = epoch + 1
        #print("Epoch number is ", epoch)
    return m, c
```

```python
# Plotting Regression Line
plt.scatter(x,y)
plt.show()

m, c = weight_update(x, y)

print('m = %s c = %s' %(m, c))

print("Regression plot after finding intercept and slope")

plt.scatter(x,y)
plt.plot(x, x*m+c, 'r')
plt.show()
```
